## 10 arms
### uniform

```python
import numpy as np
```

```python
# data generation 10 arm, uniform
np.random.seed(0)
K = 10
mean = np.random.rand(K)
#data
data = []
noise = []
Trs = 10
for i in range(Trs):
    data.append(np.random.choice(K, 1000000)) # shape(1e6, )
    noise.append(np.random.normal(0, 1, 1000000)) # shape (1e6, )
```

```python
T = 10000
sigma = 0.1
delta = 0.01
```

```python
B_list = [10, 100]
for B in B_list:
    print("B = ", B)
    L_trials = []
    error_list = []
    for trial in range(Trs):
        Y = np.zeros(K)
        loss_list, error_s = [], []
        L = 0
        Mtk = np.zeros(K)
        Ntk = np.zeros(K)
        for t in range(T):
            x = data[trial][t]
            f = mean[x]
            y = f + noise[trial][t] * sigma
            Mtk[x] +=1 
#                 print(np.sqrt(2*(sigma**2)*np.log(2*((Ntk.sum())**3)/delta)/max(Ntk[x],1)), (B**(1./(3.))) * ((t+1)**(-1./(3.))), C * (B**(1./(3.))) * ((t+1)**(-1./(3.))) )
#             if Ntk[x] == 0 or np.sqrt(2*(sigma**2)*np.log(2*((Ntk.sum())**3)/delta)/max(Ntk[x],1)) > C * (B**(1./(3.))) * ((t+1)**(-1./(3.))):
#             if Ntk[x] == 0 or np.random.rand() < 0.5:
            Ntk[x] += 1
            L += B
            Y[x] += y
            p = Y[x] / Ntk[x]
            error = abs(p - f)
            L += error
            error_s.append(error)
            loss_list.append(L/(t+1))
        error_list.append(error_s)
        L_trials.append(loss_list)
        print("trial %d total pullls: %d"%(trial, Ntk.sum()))
    np.save("./non_active/arm_%d_loss_B_%d_T_%d_uniform"%(K, B, T), L_trials)
    np.save("./non_active/arm_%d_error_B_%d_T_%d_uniform"%(K, B, T), error_list)
```
